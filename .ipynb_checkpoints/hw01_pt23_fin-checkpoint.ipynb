```python
from sklearn.cluster import KMeans
import pandas as pd
import matplotlib.pyplot as plt 
from sklearn import preprocessing
import numpy as np
```

## 2a) K-means, Hierarchical, GMM

```python
######## subset df and convert to np array ########

dataset = pd.read_csv('listings.csv')
dataset['price'] = dataset['price'].replace('[\$,]', '', regex=True).astype(float)
# subset clustering col
df = dataset.loc[:,["longitude","latitude","price"]].values
#dataset_rmna = dataset.dropna(axis=0,how='any')

print(df.shape)

```

```python
# chck for na
a = dataset["longitude"].isna().sum()
b = dataset["latitude"].isna().sum()
c = dataset["price"].isna().sum()
print(a,b,c)
```

```python
######## scale data ########
min_max_scaler = preprocessing.MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(df)
dfs = pd.DataFrame(x_scaled)
print(dfs)
```

```python
wcss = []
for i in range(1, 11): 
    kmeans = KMeans(n_clusters = i, init = 'k-means++', max_iter = 300, n_init = 10, random_state = 0)
    kmeans.fit(dfs)
    wcss.append(kmeans.inertia_)
plt.plot(range(1,11), wcss)
plt.title("Elbow method applied to Determine Optimal Cluster Number")
plt.xlabel("number of clusters")
plt.ylabel("within cluster sum of squares")
plt.show()
```

```python
# Apply kmeans to the dataset
kmeans = KMeans(n_clusters = 3, max_iter = 300, init = 'k-means++', n_init = 10, random_state = 0)
y_kmeans = kmeans.fit_predict(dfs) #predict which cluster each point belongs to
y_kmeans
```

## 3b) Pros and Cons of Clustering Algorithms

Pros of K-means:
- computationally efficient
- produce tighter clusters than hierarchical clustering

Cons of k-means:
- has no measure of probability or uncertainty of cluster assignments
- it is sensitive to outliers and initial seeds
- K (clusters) value is unknown. Although elbow method or domain knowledge can be used to determine K, this uncertainty can negatively impact clustering outcomes
- K-means also does not perform wel with clusters of different density. 


Pros of Hierarchical clustering: 
- able to expose the hierarchy in data, can demonstrate information such as the divergent evolution of species
- easier to decide on the number of clusters from the dendrogram

Cons of Hierarchical clustering:
- hierarchical clustering is computationally expensive
- the algorithm can't go back and undo what's not done correctly
- sensitive to outliers and noises

Pros of GMM:
- very fast algorithm
- addresses that the clustering assignment for some points might be more certain than others


Cons of GMM:
- if the data has insufficient many points per cluster, it is difficult to estimate the covariance matrices and the algorithm can diverge and find solutions with infinite likelihood.

```python
import folium
from folium import plugins
from folium.plugins import HeatMap
```

```python
df = pd.read_csv('listings.csv')
df = df[['latitude', 'longitude' , 'price']]
df['price'] = df['price'].replace('[\$,]', '', regex=True).astype(float)
```

```python
def generateBaseMap(default_location=[40.693943, -73.985880]):
    base_map = folium.Map(location=default_location) 
    return base_map
```

```python
base_map = generateBaseMap()
HeatMap(data=df[['latitude', 'longitude' , 'price']].
groupby(['latitude', 'longitude']).mean().
reset_index().values.tolist(), radius=8, max_zoom =13).add_to(base_map) 
base_map.save('index.html')
```
